```python
import pandas as pd

datasets = {
    dataset:pd.read_csv("./data/"+dataset+".csv") for dataset in ("Psy","KatyPerry", "LMFAO","Eminem","Shakira")
}
```

### Descrição dos atributos
<div style="text-align: justify">
A coleção é composta por um arquivo CSV por conjunto de dados, o qual segue o seguinte formato:
</div><br>

| COMMENT_ID | AUTHOR | DATE | CONTENT | CLASS |
|:-------------------------------------:|:---------------:|:-------------------:|:------------------------:|:---:|
| z12oglnpoq3gjh4om04cfdlbgp2uepyytpw0k | Francisco Nora, | 2013-11-28T19:52:35 | please like :D [weblink] | 1 |

<br>
<div style="text-align: justify">
Os atributos mais interessantes aqui são o conteúdo do comentário, obviamente, mas também a data de publicação do comentário. Com a data podemos investigar se existe um momento após a publicação do vídeo onde os comentários do tipo spam tendem a ser mais publicados. Para tal, precisamos converter a data no formato de string para um valor quatitativo real. Outro problema encontrado na data foram os formatos diferentes, com ou sem micro segundos, e a falta de data de publicação de alguns comentários. Para resolver o primeiro problema optamos por eliminar os micro segundos. Já a falta de valores foram contornador gerando datas artificias repetindo a data conhecida mais proxima, tendo em vista que, segundo o artigo, as publicações preservam a ordem cronologica. Primeiro vamos resolver os problemas antes de normalizar a partir de uma época.
</div>


```python
import time
from calendar import timegm

def solveDateProblems(dates):
    dates = list(dates)
    
    for i in range(len(dates)):
        date = str(dates[i])
        closestDate = ''
        
        if date != 'nan':
            closestDate = date.partition(".")[0] # removendo os micro segundos   
        
        if closestDate != '':
            for j in range(i-1,-1,-1):
                if str(dates[j])!='nan': break
                dates[j] = closestDate
            dates[i] = closestDate
            
    return pd.Series(dates)

def datesSinceAnEpoch(dates):
    return dates.apply(lambda date: timegm(time.strptime(date, "%Y-%m-%dT%H:%M:%S")))

for key in datasets.keys():
    noProblemDates = solveDateProblems(datasets[key].DATE)
    datasets[key].DATE = datesSinceAnEpoch(noProblemDates)
    
```

Agora com as datas normalizadas podemos fazer um plot das datas de cada comentário
